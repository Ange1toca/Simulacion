#Problema 2: Distribución Exponencial
**Contexto**: Una estación de carga rápida para vehículos eléctricos quiere
simular el tiempo entre la llegada de los vehículos para prever la cantidad
de puntos de carga necesarios durante el día. La llegada de los vehículos
sigue una distribución exponencial con una tasa de 1 llegada cada 15
minutos, en promedio.

**Objetivo**: Generar tiempos entre llegadas de vehículos eléctricos durante
un día de operación (12 horas)

##Jose Angel Toca Zamora

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (tasa de llegada)
lambda_llegada = 1 / 15  # Un vehículo cada 15 minutos en promedio

def generar_tiempo_entre_llegadas(lambda_llegada):
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    tiempo_entre_llegadas = -1 / lambda_llegada * np.log(1 - u)  # Aplicar transformada inversa
    return tiempo_entre_llegadas

# Simular llegadas durante 12 horas (720 minutos)
tiempo_total = 0
llegadas = []
while tiempo_total < 720:
    tiempo_entre = generar_tiempo_entre_llegadas(lambda_llegada)
    tiempo_total += tiempo_entre
    if tiempo_total < 720:
        llegadas.append(tiempo_entre)

# Mostrar los tiempos entre llegadas
print("Tiempos entre llegadas de vehículos generados:", llegadas)

# Graficar los tiempos entre llegadas simulados
plt.hist(llegadas, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos Entre Llegadas (Exponencial)')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

# **Conclusion**
La simulación de los tiempos entre llegadas de vehículos eléctricos en una estación de carga rápida utilizando la distribución exponencial nos permite comprender mejor el comportamiento de la demanda y planificar la capacidad de carga de manera eficiente.

 la simulación mediante la distribución exponencial es una herramienta valiosa para la gestión de la demanda en estaciones de carga rápida, permitiendo mejorar la eficiencia operativa y optimizar la atención a los usuarios.
