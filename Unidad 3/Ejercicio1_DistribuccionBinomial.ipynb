#Problema 1: Distribución Binomial
**Contexto**: Un distribuidor de artículos electrónicos recibe varios pedidos
diarios. Cada día, el distribuidor tiene una probabilidad de éxito de cumplir
un pedido (es decir, de tener en stock los artículos solicitados) del 70%.
Desea simular la cantidad de pedidos diarios que logra satisfacer para
gestionar adecuadamente sus inventarios y minimizar las faltantes de
stock.

**Objetivo**: Generar una variable aleatoria que simule el número de pedidos
satisfechos por día durante una semana, considerando que cada día el
distribuidor recibe 20 pedidos.

##Jose Angel Toca Zamora


```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

# Parámetros de la distribución binomial
n = 20   # Número de pedidos diarios
p = 0.70 # Probabilidad de satisfacer un pedido

def binomial_pmf(n, k, p):
    return comb(n, k) * (p**k) * ((1 - p)**(n - k))

def binomial_inverse_transform(n, p, U):
    F_k = 0  # Inicializamos la FDA
    k = 0
    while True:
        F_k += binomial_pmf(n, k, p)  # Sumamos la probabilidad hasta que F(k) >= U
        if U <= F_k:
            return k  # Este valor de k es el número de pedidos satisfechos
        k += 1

# Número de días a simular
num_dias = 7

# Generar números aleatorios uniformes
U = np.random.uniform(0, 1, num_dias)

# Simular la cantidad de pedidos satisfechos por día
pedidos_satisfechos = [binomial_inverse_transform(n, p, u) for u in U]

# Mostrar resultados
print("Número de pedidos satisfechos por día:")
print(pedidos_satisfechos)

# Graficar los resultados
plt.hist(pedidos_satisfechos, bins=np.arange(min(pedidos_satisfechos), max(pedidos_satisfechos) + 1) - 0.5, edgecolor='black')
plt.title("Simulación de pedidos satisfechos por día (Distribución Binomial)")
plt.xlabel("Número de pedidos satisfechos")
plt.ylabel("Frecuencia")
plt.show()
```

# **CONCLUSION**
La simulación realizada utilizando la distribución binomial nos permite analizar el desempeño del distribuidor en términos de satisfacción de pedidos diarios. A través de este modelo probabilístico, se observó que la cantidad de pedidos satisfechos varía debido a la incertidumbre inherente en el proceso de disponibilidad de stock.

 la simulación con variables aleatorias y la distribución binomial resulta una herramienta valiosa para la gestión de inventarios y la optimización de procesos logísticos, permitiendo reducir la incertidumbre y mejorar la eficiencia operativa del distribuidor.
